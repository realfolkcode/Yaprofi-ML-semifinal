```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('data.csv', header=None)
```

```python
data
```

```python
X = pd.DataFrame(data[0])
y = data[1]
```

```python
X
```

```python
y
```

```python
X = X.rename(columns={0: 'x'})
```

```python
X
```

```python
X['s2'] = np.sin(X['x'])**2
```

```python
X['sl'] = np.sin(X['x']) * np.log(X['x'])
```

```python
X['l2'] = np.log(X['x'])**2
```

```python
X['x2'] = X['x']**2
```

```python
X
```

```python
X = X.drop(columns=['x'])
```

```python
X
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
model.coef_
```

```python
a = np.sqrt(model.coef_[0])
b = np.sqrt(model.coef_[2])
c = model.coef_[3]
```
